```python
import cv2 # opencv für Bildbeareitung
import matplotlib.pyplot as plt # pyplot um Bild anzuzeigen

bild = cv2.imread("bild.jpg") # bild lesen im BGR-Format

bild = cv2.cvtColor(bild, cv2.COLOR_BGR2RGB) # von BGR in RGB umwandeln

plt.imshow(bild) # Bild anzeigen im RGB-Format
```

```python
graustufen = cv2.cvtColor(bild, cv2.COLOR_RGB2GRAY) # von RGB in Graustufen umwandeln
plt.imshow(graustufen, cmap="gray")
```
